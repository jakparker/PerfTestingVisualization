```python
import pandas
from IPython.display import display

def shiftTimeStamp(row):
    return row['timeStamp'] - 1688695066761


df = pandas.read_csv('logs_dress_all2.csv')
df['modTimeStamp'] = df.apply(shiftTimeStamp, axis=1)
display(df)
```
